```python
orginals_sequels = sequels.merge(sequels, left_on = 'id', right_on = 'sequel', suffixes = ('_seq','_org'))
orginals_sequels[['title_org','title_seq']]
```

```python
file_name10 = 'crews.p'
path10 = os.path.join(grandparent_path,data_dir_name,file_name10)
crews = pd.read_pickle(path10)
print(crews.shape)
crews.head()
```

```python
boolean_filter = ((crews_self_merged['job_dir'] == 'Director')&(crews_self_merged['job_crew'] != 'Director'))
dir_crews = crews_self_merged[boolean_filter]
dir_crews.head()
```
